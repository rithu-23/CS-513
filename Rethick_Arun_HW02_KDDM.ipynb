```python
#Importing Required Libraries For EDA Analysis
import pandas 
import numpy
import matplotlib.pyplot as plot
import seaborn
```

```python
#Loading given dataset into a pandas dataframe, and printing the summary,first and least 10 data samples
dataFrame = pandas.read_csv('/Users/rithu/Downloads/breast-cancer-wisconsin.csv') #Path of given dataset(csv) file in my system,for 
                                                                                  #testing and grading,please use appropriate path
print("Summary of Data loaded into Data Frame : \n")
print(dataFrame.describe())
print("\nFirst Ten Data Sample : \n")
print(dataFrame.head(10))
print("\nLast Ten Data Sample : \n")
print(dataFrame.tail(10))
```

```python
#Summarizing each column (min,max and mean) - Asked in part I. of the assignment
print("\nMin,Max and Mean : \n")
summaryDataFrame = dataFrame.describe().T[["min","max","mean"]]#Transposing the output for better readability
print(summaryDataFrame)
```

```python
#Identifying Missing Values - part II. of the assignment
dataFrame['F6'] = pandas.to_numeric(dataFrame['F6'], errors='coerce')#We can see in the previous table that F6 is missing. This is because F6
                                                                     #is not numeric(int or float). Therefore,converting F6 to numeric type.
print("\nMissing values : \n")
print(dataFrame[dataFrame.isna().any(axis=1)])#Print the missing values
missingValues = dataFrame.isnull().sum()#Identifying Count Of Missing Values
print("\nMissing values per column : \n")
print(missingValues)#Print the count of missing values
```

```python
#Replacing Missing Values with Mean of the Column - part III. of the assignment
meanF6 = dataFrame['F6'].mean()
print("Mean of Column F6 is " + str(meanF6))
dataFrame.fillna({'F6': dataFrame['F6'].mean()}, inplace=True)
missingValues = dataFrame.isnull().sum()#Recheck Count Of Missing Values
print("\nMissing values per column after replacing missing values with Mean: \n")
print(missingValues)#Rerint the count of missing values
print("\nSummary of the dataset after replacing null values with mean\n")
print(dataFrame.describe())#Summary of Dataset after replacing null values with mean
print("\nMin, Max and Mean after replacing null values with mean\n")
summaryDataFrame = dataFrame.describe().T[["min","max","mean"]]#Transposing the output for better readability
print(summaryDataFrame)
```

```python
#Displaying Frequency Table of Class vs F6 - part IV. of the assignments
frequencyTable = pandas.crosstab(dataFrame['F6'],dataFrame['Class'])#Inverted Class and F6 for better readability
print(frequencyTable)#Print out the frequency table
```
